# EXPLORATORY DATA ANALYSIS

what is the overall sales trend?

```python
mydata['order_date'].min()
```

```python
mydata['order_date'].max()
```

```python
mydata['month_year'] = mydata['order_date'].apply(lambda x: x.strftime('%y-%m') )
```

```python
# grouping month year
mydata_trend = mydata.groupby('month_year').sum()['sales'].reset_index()
```

```python
# visualize the sales tend 
plt.figure(figsize = (15, 6))
plt.plot(mydata_trend['month_year'], mydata_trend['sales'], color = 'red')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

what are the top 10 products by sales?

```python
mydata.head(5)
```

```python
mydata['product_name']
```

```python
# creating the dataframe of product_name and sales 
prod_sales = pd.DataFrame(mydata.groupby('product_name').sum()['sales'])
```

```python
# sort the sales in ascending order.
prod_sales = prod_sales.sort_values('sales', ascending=False)
```

```python
prod_sales[:10]
```
